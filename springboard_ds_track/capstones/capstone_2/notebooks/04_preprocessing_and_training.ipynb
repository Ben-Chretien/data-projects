# Capstone 2: Modeling Preparation and Preprocessing

## Overview

This notebook prepares the dataset for modeling by defining the target variable, selecting relevant features, and transforming variables into a format suitable for machine learning.

## 1. Data Loading and Inspection

```python
# Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# Load dataset
df = pd.read_csv('../data/airline_delay_cause.csv')
```

```python
# Quick inspection
print(df.head().T)
print(df.info())
```

## 2. Feature Selection

Select features based on EDA findings, focusing on flight volume, seasonality, and airport and airline effects.

```python
# Select features based on EDA findings
# These represent volume, seasonality, and location effects
X = df[['arr_flights', 'month', 'carrier', 'airport']]
```

## 3. Target Variable Definition

```python
# Define the target variable
y = df['arr_delay']
```

## 4. Leakage Considerations

Delay component variables are excluded because they directly sum to the target variable (`arr_delay`), which would introduce data leakage.

## 5. Categorical Encoding

```python
# Convert categorical variables into numerical format
X = pd.get_dummies(X, columns=['carrier', 'airport'], drop_first=True)
```

## 6. Preprocessing Check

```python
print(X.head())
print(X.shape)
```

## 7. Feature Scaling Considerations

Feature scaling was not applied, as most features are one-hot encoded and already in a consistent format. Scaling may be considered later depending on the model used.

## 8. Train-Test Split

```python
# Split data into training and testing sets
# Training data is used to build the model
# Testing data is used to evaluate performance
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1)

# Check resulting shapes
X_train.shape, X_test.shape
```

## 9. Next Steps / Conclusion  

The dataset has been prepared for modeling by selecting relevant features, defining the target variable, encoding categorical variables, and splitting the data into training and testing sets. Feature scaling will be applied in the modeling stage after the split to avoid data leakage.
